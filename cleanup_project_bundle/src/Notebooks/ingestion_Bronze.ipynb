```python
%sql
create connection if not exists earthquake_conn
type HTTP
options(
  host = 'https://earthquake.usgs.gov',
  port = 443,
  base_path = '/earthquakes/feed/v1.0/',
  bearer_token = 'na'
)
```

```python
from databricks.sdk import WorkspaceClient

w=WorkspaceClient()
conn=w.connections.get("earthquake_conn")
base_url = f"{conn.options['host']}{conn.options['base_path']}"
```

```python
dbutils.widgets.text('catalog_name', 'cleanup_cat', 'cleanup_cat')
catalog_name = dbutils.widgets.get('catalog_name')
```

```python
spark.sql(
    f"use catalog {catalog_name/Volumes/cleanup_cat/bronze/earthquake_data/Volumes/cleanup_cat/bronze/earthquake_data/Volumes/cleanup_cat/bronze/earthquake_data/Volumes/cleanup_cat/bronze/earthquake_data/earthquake_data_2026-04-06.json}"
)

spark.sql(
    "use schema bronze"
)

spark.sql(
    "create volume if not exists earthquake_data"
)
```

```python
import requests
import json
import datetime
url=f'{base_url}summary/all_day.geojson'
response=requests.get(url)
data=response.json()
if response.status_code!=200:
  raise Exception(f'error in getting data from {url}')


current_date=datetime.datetime.now().strftime('%Y-%m-%d')
dbutils.fs.put(f'/Volumes/{catalog_name}/bronze/earthquake_data/earthquake_data_{current_date}.json',json.dumps(data),overwrite=True)
```
